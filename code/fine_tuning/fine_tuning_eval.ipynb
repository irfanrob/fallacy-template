```python
df = pd.read_csv('other_outputs.csv')
df
```

```python
fallacy = df['fallacy'].to_list()
input = df['input'].to_list()
output = df['output'].to_list()
gold_label = df['gold_label'].to_list()
```

```python
template_output = []
slot_A = []
slot_C = []
slot_A_prime = []
slot_C_prime = []
slot_X = []
template_gold = []
slot_A_gold = []
slot_C_gold = []
slot_A_prime_gold = []
slot_C_prime_gold = []
slot_X_gold = []
```

```python
def split_results(results,template,a,c,a_prime,c_prime,x):
  for i in range (len(results)):
    #template number
    if re.search(r'Template No.=([0-5])',results[i]):
      template_no = re.findall(r'Template No.=([0-5])',results[i])
      template.append(template_no[0])
    else:
      template_no = '5'
      template.append(template_no)
    #A
    if re.search(r'\[A\]=(.*)',results[i]):
      A = re.findall(r'\[A\]=(.*)',results[i])
      a.append(A[0])
    else:
      A = ''
      a.append(A)
    #C
    if re.search(r'\[C\]=(.*)',results[i]):
      C = re.findall(r'\[C\]=(.*)',results[i])
      c.append(C[0])
    else:
      C = ''
      c.append(C)
    #A'
    if re.search(r'\[A\'\]=(.*)',results[i]):
      A1 = re.findall(r'\[A\'\]=(.*)',results[i])
      a_prime.append(A1[0])
    else:
      A1 = ''
      a_prime.append(A1)
    #C'
    if re.search(r'\[C\'\]=(.*)',results[i]):
      C1 = re.findall(r'\[C\'\]=(.*)',results[i])
      c_prime.append(C1[0])
    else:
      C1 = ''
      c_prime.append(C1)
    #X
    if re.search(r'\[X\]=(.*)',results[i]):
      X = re.findall(r'\[X\]=(.*)',results[i])
      x.append(X[0])
    else:
      X = ''
      x.append(X)
```

```python
split_results(output,template_output,slot_A,slot_C,slot_A_prime,slot_C_prime,slot_X)
split_results(gold_label,template_gold,slot_A_gold,slot_C_gold,slot_A_prime_gold,slot_C_prime_gold,slot_X_gold)
```

```python
result_df = pd.DataFrame({
    'input': input,
    'fallacy': fallacy,
    'A' : slot_A_gold,
    'C': slot_C_gold,
    "A'": slot_A_prime_gold,
    "C'": slot_C_prime_gold,
    "X": slot_X_gold,
    "template": template_gold,
    "template_pred": template_output,
    'A_pred' : slot_A,
    'C_pred': slot_C,
    "A'_pred": slot_A_prime,
    "C'_pred": slot_C_prime,
    "X_pred": slot_X
})
```

```python
result_df = result_df[result_df['template']!="5"]
result_df = result_df[result_df['template']==result_df['template_pred']]
result_df.reset_index(drop=True, inplace=True)
```

```python
def word_overlap(text1, text2):
    set1 = set(text1.lower().split())
    set2 = set(text2.lower().split())
    intersection = set1.intersection(set2)
    overlap_ratio = len(intersection) / max(len(set1), len(set2))
    return overlap_ratio
```

```python
output_comb = []
gold_comb = []
for i in range (len(result_df)):
  if result_df['fallacy'][i]=="Fallacy of Credibility":
    output_combine = result_df['A'][i] + ' ' + result_df['C'][i] + ' ' + result_df["A'"][i] + ' ' + result_df["C'"][i] + ' ' + result_df['X'][i]
    output_comb.append(output_combine)
    gold_combine = result_df['A_pred'][i] + ' ' + result_df['C_pred'][i] + ' ' + result_df["A'_pred"][i] + ' ' + result_df["C'_pred"][i] + ' ' + result_df['X_pred'][i]
    gold_comb.append(gold_combine)
```

```python
overlap = []
for i in range (len(output_comb)):
  temp = word_overlap(output_comb[i],gold_comb[i])
  overlap.append(temp)
```

```python
overlap_50 = [k > 0.5 for k in overlap]
overlap_100 = [k == 1.0 for k in overlap]
```

```python
denominator = 0
for i in range (len(result_df)):
  if result_df['fallacy'][i]=="Fallacy of Credibility":
    denominator = denominator + 1
```

```python
total_rows = len(result_df) #accuracy total
accuracy_50 = sum(overlap_50) / denominator
accuracy_100 = sum(overlap_100) / denominator

print(f"Partial Match: {accuracy_50:.2f}")
print(f"Exact Match: {accuracy_100:.2f}")
```
